```python
sns.scatterplot(x="total",y="speeding",data=df)
```

Inference:from the plot we can say that total no.of drivers on fatal collisions is directly proportional to percentage of drivers involved in fatal collisions,who were speeding.

```python
sns.scatterplot(x='total',y='alcohol',data=df,color="r")
```

Inference:from the plot we can say that total no.of drivers on fatal collisions is directly proportional to percentage of drivers involved in fatal collisions,consuming alcohol.


```python
sns.scatterplot(x='total',y='no_previous',data=df,color='g')
```

Inference: from the plot we can say that total no.of drivers on fatal collisions is directly proportional to percentage of drivers involved in fatal collisions who are do not have previous accidents

Lineplot

```python
sns.lineplot(x="total",y="not_distracted",data=df,errorbar=None)
```

Inference involved:From the plot it is evident that the total number of drivers in fatal collisions is linearly proportional to percentage of drivers involved in fatal collisions who were not distracted.

Distribution plot

```python
sns.displot(df['ins_premium'])
```

Inference:The count is highest in between ins_premium of 350-900

Relplot

```python
sns.relplot(x='alcohol',y='speeding',data=df,hue="abbrev")
```

Inference: In the above plot we can say that with an increase of alocohol intake speed also increases.

Jointplot

```python
sns.jointplot(x="ins_premium",y="ins_losses",data=df)
```

Inference: From the above plot we can say that ins_premium and ins_losses are directly proportional to each other

Barplot

```python
plt.figure(figsize=(20,18))
sns.barplot(data=df,x="ins_premium",y="total")
plt.show()
```

Inference:Most of the ins_premium that have high no.of drivers in fatal collisions is in the range of 669,871,992

Boxplot

```python
plt.figure(figsize=(20,18))
sns.boxplot(x=df["total"],y=df["alcohol"],data=df)
plt.show()
```

Inference: From the boxplot we can say that there are no outliers
